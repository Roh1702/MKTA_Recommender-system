```python
##### Recommend Items to our Customer based on other similar customer purchased history
# df =>  Raw data (406829 rows × 8 columns) with Description
# df_main => Boolean Matrix of Customer ID & Stock Codes
# Cosine_simi_mat => Cosine similarity matrix of Customer ID

cus_id = input("Enter Your Customer ID: ") # To ask User Customer ID


if float(cus_id) not in cosine_simi_mat.columns:
    print("Customer ID",cus_id, "is Incorect") # Condition to check Valid customer in our Dataset
    
else:
    my_customer = cosine_simi_mat.loc[float(cus_id),:].sort_values(ascending=False).head(3) # checking Top 5 similar customer order by descending
    similar_customer = my_customer.index        # list of top 5 similar customer
    df2 = df_main.loc[similar_customer,:]       # Create dataframe for similar customer
    df3 = df2.loc[:, (df2 != 0).any(axis=0)]    # Dropling those columns/Items which is not brought by any of the similar customer
    df4 = df3.iloc[[0,]]                        # Checking which product not bought by a my_customer to recommend him/her
    df5 = df4.loc[:, (df4 !=1).any(axis=0)]     # Checking which product not bought by a my_customer to recommend him/her
    print(f"Total No of recommended Items : {len(df5.columns)}")  # printing total no of item to recommend
    print("Recommended Items are :")            # Printing list of item "StockCode" to recommend
    # To diplay Name of Recommended Items to  my_customer 
    L = []    
    for i in df5.columns:
        L.append(np.min(df.index[df['StockCode'] == str(i) ]))
        
    df6 = df.loc[L,['StockCode','Description']]

    print(df6)
```
